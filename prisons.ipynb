```python
import matplotlib.pyplot as plt

x = male_79['Most Serious Crime'] 
y = male_79['Age at Admission']
plt.plot(x,y,'ro',label = 'Crimes')
plt.title('Crimes > 79')
plt.xlabel('Crimes')
plt.xticks(rotation=90)
plt.ylabel('Ages')
yint = range(min(y),(max(y))+1,5)
plt.yticks(yint)
plt.grid('True')
plt.legend(loc = 'best')
plt.show()
```
