```python
#let us rebuild the model with all the features of the table titanic
from sklearn.linear_model import LogisticRegression

X=titanic[['Pclass','male','Age','Siblings/Spouses','Parents/Children','Fare']].values
y=titanic['Survived'].values
model=LogisticRegression()
model.fit(X,y)
model.predict(X)
```

```python
model.predict([[3,True,22.0,1,0,7.25]])
#this passenger did not survived
```

```python
model.predict(X[:5])
```

```python
y[:5]
#prediction against reallity are all correct
```

```python
#Score the model.we can count the number od datapoints it predicts correctly. Accuracy score
#let us create an array that has predicted y values

y_pred =model.predict(X)
y_pred
```

```python
#our accuracy score
(100*((y_pred==y).sum())/y.shape[0]).round(2)
#makes the correct prediction on 80% of the datapoints
```
